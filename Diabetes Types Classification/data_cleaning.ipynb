```python
# Imports

import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
# Loading the dataset

df = pd.read_csv('./Datasets/data.csv')
df.head()
```

```python
df.shape
```

```python
df.columns.to_list()
```

```python
df = df[['Age','BMI','Insulin Levels','Blood Pressure','Blood Glucose Levels','Physical Activity','Smoking Status',
 'Alcohol Consumption','Family History','Target']]
```

```python
df
```

```python
df['Target'].unique()
```

```python
chosen_targets = ['Prediabetic','Type 1 Diabetes','Type 2 Diabetes']

df1 = df[df['Target'].isin(chosen_targets)].copy()

df1
```

```python
df1.dropna(inplace=True)
df1.drop_duplicates(inplace=True)
```

```python
df1.info()
```

```python
df1['Smoking Status'].unique()
```

```python
df1['Smoking Status'] = df1['Smoking Status'].map({'Smoker':True, 'Non-Smoker':False}).astype(bool)
```

```python
df1['Family History'].unique()
```

```python
df1['Family History'] = df1['Family History'].map({'Yes':True, 'No':False}).astype(bool)
```

```python
df1['Physical Activity'].unique()
```

```python
df1['Alcohol Consumption'].unique()
```

```python
df2 = pd.get_dummies(df1, columns=['Alcohol Consumption', 'Physical Activity']).copy()
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```
