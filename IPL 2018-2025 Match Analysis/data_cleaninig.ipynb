# Loading the IPL_2025 file for data cleaning and processing


```python
import pandas as pd
pd.set_option('display.max_colwidth', 30)
```

```python
path = "./Scraped Data/ipl_2018.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0': 'match_number'}).astype({'match_number': 'int'})
```

```python
df['match_number'] = df['match_number'] + 1
df.set_index('match_number', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
null_rows = df[df.isnull().any(axis=1)]
null_rows
```

```python
df["first_inn_team"] = df["first_inn_team"].str.split(r"\d", n=1).str[0].str.strip()
df["second_inn_team"] = df["second_inn_team"].str.split(r"\d", n=1).str[0].str.strip()
```

```python
df.head()
```

```python
df['first_inn_team'].unique()
```

```python
team_map = {
    "Mumbai Indians": "MI",
    "Chennai Super Kings": "CSK",
    "Kolkata Knight Riders": "KKR",
    "Delhi Capitals": "DC",
    "Sunrisers Hyderabad": "SRH",
    "Rajasthan Royals": "RR",
    "Lucknow Super Giants": "LSG",
    "Punjab Kings": "PBKS",
    "Gujarat Titans": "GT"
}
df['first_inn_team'] = df['first_inn_team'].replace(team_map)
df['second_inn_team'] = df['second_inn_team'].replace(team_map)
```

```python
df['first_inn_team'].unique()
```

```python
df.head()
```

```python
df.info()
```

```python
df['match_date'] = df['match_date'].str.split(',').str[0].str.strip()
```

```python
df.head()
```

```python
df.to_csv('ipl_2018_cleaned.csv')
```
