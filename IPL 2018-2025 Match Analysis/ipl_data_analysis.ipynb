# Loading IPL Data

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ipl_2018-2025.csv', index_col=0)
```

```python
df.head()
```

```python
total_matches = len(df)
f"Total matches played: {total_matches}"
```

```python
df.info()
```

```python
df.drop(['match_link'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
df.head()
```

```python
team_map = {
    "Mumbai Indians": "MI",
    "Chennai Super Kings": "CSK",
    "Kolkata Knight Riders": "KKR",
    "Royal Challengers Bengaluru": "RCB",
    "Delhi Capitals": "DC",
    "Sunrisers Hyderabad": "SRH",
    "Rajasthan Royals": "RR",
    "Lucknow Super Giants": "LSG",
    "Punjab Kings": "PBKS",
    "Gujarat Titans": "GT"
}


df['match_result'] = df['match_result'].replace(team_map, regex=True)
```

```python
df
```

```python
# Distribution of matches per year
df['year'].value_counts().sort_index()
```

```python
# Unique teams 

df['first_inn_team'].value_counts()
```

```python
df['second_inn_team'].value_counts()
```

```python
total_matches_played = {}
for _, row in df.iterrows():
    first_team = row['first_inn_team']
    second_team = row['second_inn_team']
    total_matches_played[first_team] = total_matches_played.get(first_team, 0) + 1
    total_matches_played[second_team] = total_matches_played.get(second_team, 0) + 1
    
df1 = pd.DataFrame(list(total_matches_played.items()), columns=['team', 'played'])
df1
```

```python
total_wins = {}
for _, row in df.iterrows():
    result = row['match_result']
    if 'won' in result:
        if 'Match tied (' in result:
            # Extract winner from tied match format: "Match tied (TEAM won...)"
            winner = result.split('Match tied (')[1].split(' won')[0]
        else:
            # Regular win format: "TEAM won by..."
            winner = result.split(' won')[0]
        total_wins[winner] = total_wins.get(winner, 0) + 1
        
    
df2 = pd.DataFrame(list(total_wins.items()), columns=['team', 'won'])
```

```python
df2
```

```python
total_losses = {}
for _, row in df.iterrows():
    result = row['match_result']
    if 'won' in result:
        if 'Match tied (' in result:
            # Extract winner from tied match format: "Match tied (TEAM won...)"
            winner = result.split('Match tied (')[1].split(' won')[0]
        else:
            # Regular win format: "TEAM won by..."
            winner = result.split(' won')[0]
        # Determine loser
        first_team = row['first_inn_team']
        second_team = row['second_inn_team']
        loser = second_team if winner == first_team else first_team
        total_losses[loser] = total_losses.get(loser, 0) + 1
        
        
df3 = pd.DataFrame(list(total_losses.items()), columns=['team', 'losses'])
df3
```

```python
all_teams = set()
for col in ['first_inn_team', 'second_inn_team']:
    all_teams.update(df[col].unique())
    
no_result_dict = {team: 0 for team in all_teams}
for _, row in df.iterrows():
    result = row['match_result']
    if 'no result' in result.lower() or 'abandoned' in result.lower():
        first_team = row['first_inn_team']
        second_team = row['second_inn_team']
        no_result_dict[first_team] += 1
        no_result_dict[second_team] += 1
        
        
df4 = pd.DataFrame(list(no_result_dict.items()), columns=['team', 'no_results'])
df4
```

```python
df1 = df1.merge(df2, on='team')
df1 = df1.merge(df3, on='team')
df1 = df1.merge(df4, on='team')
```

```python
df1 = df1.copy()
```

```python
df1
```

```python
df1.sort_values(by='won', ascending=False, inplace=True)
df1.reset_index(drop=True, inplace=True)
df1
```

# EDA

```python
# plotting a bar chart for team and it's winings

ax = sns.barplot(data=df1, x='team', y='won',hue='played', palette='coolwarm')

sns.set_theme(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for team and it's losses

ax = sns.barplot(data=df1, x='team', y='losses',hue='played', palette='coolwarm')

sns.set_theme(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
